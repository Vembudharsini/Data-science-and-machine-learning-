```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#bar chart 
a = np.array([1,2,3,4,5])
b = np.array([50,60,70,90,100])
plt.bar(a,b)
plt.xlabel("day")
plt.ylabel("sales")
plt.title("Day vs Sales")
plt.show()
```

```python
#line chart 
a = np.array([1,2,3,4,5])
b = np.array([50,60,70,80,90])
plt.plot(a,b)
plt.plot(a,b,marker = 'o',ms = 20)
#plt.plot(a,b,marker = 'd',ms = 20)      
#plt.plot(a,b,marker = 'h',ms = 20)
#plt.plot(a,b,marker = 'H',ms = 20)
##plt.plot(a,b,marker = '^',ms = 20)
plt.xlabel("day")
plt.ylabel("sales")
plt.title("Day vs Sales")
plt.show()
```

```python
#scatter 
a = np.array([1,2,3,4,5])
b = np.array([50,60,70,90,100])
plt.scatter(a,b)
plt.xlabel("day")
plt.ylabel("sales")
plt.title("Day vs Sales")
plt.show()
```

```python
#histogram
b = np.array([50,52,60,66,70,90,100])
plt.hist(b)
plt.xlabel("day")
plt.ylabel("sales")
plt.title("Day vs Sales")
plt.show()
```

```python
#pie chart 
region=["west","north","east","south"]
sales = [23,27,22,29]
plt.pie(sales,labels=region,autopct = '%1.1f%%')
#or
#plt.pie(sales,labels=region,autopct = '%1.1f%%',colors=["pink","blue","pink","orange"])
plt.legend()
plt.show()
#or
'''import matplotlib.pyplot as plt
import numpy as np

region = ["west", "north", "east", "south"]
sales = [23, 27, 22, 29]

cmap = plt.cm.viridis
colors = cmap(np.linspace(0, 1, len(sales)))

plt.pie(
    sales,
    labels=region,
    autopct='%1.1f%%',
    colors=colors
)

plt.legend()
plt.show()'''
```

```python
#subplot
a=np.array([1,2,3,4,5])
b=np.array([50,60,70,80,90])
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)

plt.bar(a,b)
plt.subplot(1,2,2)
plt.plot(a,b)

plt.title("days vs sales")
plt.xlabel("days")
plt.ylabel("sales")
plt.show()
```
